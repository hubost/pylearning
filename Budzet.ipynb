```python
dataframe['Nowa wartość'] = dataframe.apply(lambda row: row['Wartość'] if row['Rodzaj'] == 'Przychód' else row['Wartość'] * -1, axis=1)
```

```python
dataframe['Nowa wartość'].sum()
```

```python
dataframe.groupby([pd.Grouper(key='Data', freq='Y'), pd.Grouper(key='Data', freq='M')])['Nowa wartość'].sum()
```

```python
dataframe['Rok'] = dataframe['Data'].dt.strftime('%Y')
dataframe['Miesiąc'] = dataframe['Data'].dt.strftime('%m')
dataframe.groupby(['Rok','Miesiąc'])['Nowa wartość'].sum()
```

```python
grupowanie= dataframe.groupby(['Rok','Miesiąc'])['Nowa wartość'].sum()
```
